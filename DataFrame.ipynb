```python
np.random.seed(100)
df = pd.DataFrame(np.random.randn(5,4),['A','B','C','D','E'],['W','X','Y','Z'])
df>0
```

```python
df['W']>0
```

```python
#wherever condition is true it will display value
df[df>0]
```

```python
df['W']>0
```

```python
df[df['W']>0]
# as only 2nd value was true so that row is dislayed
```

```python
#Display value of only column Y which is greater than Zero
df[df['Y']>0]['Y']
```

```python
#Display only column values of X & Z if both are greater than Zero
df[(df['X']>0) & (df['Z']>0)] [['X','Z']]
```

```python
#Display only values of W & Z if either of them is >0
df[(df['W']>0) | (df['Z']>0)] [['W','Z']]
```

```python
#Display only values of W & Z if either of them is >0; display only row C
df[(df['W']>0) | (df['Z']>0)] [['W','Z']].loc['C']
```
