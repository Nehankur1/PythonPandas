```python
import numpy as np
import pandas as pd
df = pd.DataFrame({
    'col1': [1,2,3,4],
    'col2': [444,555,666,444],
    'col3': ['abc','def','ijk','xyz']
})
df
```

```python
df['col2'].unique
```

```python
# Display value count
df['col2'].value_counts()
```

```python
#Apply funstion
def times2(x):
    return x*2
df['col1'].apply(times2)
```

```python
#Apply via Lambda function
df['col2'].apply(lambda x:x*2)
```
