Step 1: Create a Pivot table with 'product' as the rows, 'month' as the columns, and 'revenue' as the values.

Step 2: use the 'sum' aggregation function to sum the revenues for each product and month.

Step 3: Sort the pivot table by the sum of revenues in descending order using the sort_values() function.

Step 4: Extract the product with the highest revenue and assign it to the variable 'highest_revenue_product' using the index() function to select the first row, also extract the product with the second highest revenue and assign it to the variable 'second_highest_revenue_product' using the index() function to select the second row.

Step 5: Print out the names of the two products.

```python
import pandas as pd

# create a hardcoded Pandas DataFrame
df = pd.DataFrame({'product': ['A', 'A','B', 'C', 'D', 'E', 'A', 'B', 'C', 'D', 'E'],
                   'month': ['Jan', 'Jan','Jan', 'Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Feb', 'Feb'],
                   'revenue': [100, 300,200, 150, 300, 250, 150, 250, 200, 350, 300]})

# create a pivot table with 'product' as the rows and 'month' as the columns, with 'revenue' as the values
pivot = pd.pivot_table(df, index='product', columns='month', values='revenue', aggfunc='sum')
pivot
```

```python
pivot['total'] = pivot['Feb'] + pivot['Jan']
pivot
```

```python
# sort the pivot table by the sum of revenues in descending order
pivot = pivot.sort_values('total', ascending=False)
pivot
```

```python
# extract the product with the highest revenue and assign it to the variable 'highest_revenue_product'
highest_revenue_product = pivot.index[0]

# extract the product with the second highest revenue and assign it to the variable 'second_highest_revenue_product'
second_highest_revenue_product = pivot.index[1]

print("The product with the highest revenue is", highest_revenue_product)
print("The product with the second highest revenue is", second_highest_revenue_product)
```
