```python
#Applying Groupby on dataframe
import pandas as pd
data= {
    'Company': ["Airtel","Vodafone","Jio","Idea","Airtel","Jio"],
     'Person': ["Agent1","Agent2","Agent3","Agent4","Agent5","Agent6"],
     'Sales' : [100,120,130,150,100,76]
}
df = pd.DataFrame(data)
df
```

```python
df1 = df.groupby(['Company','Person','Sales']).sum()
df1
```

```python
#Joins
data1= {
    'Organization': ["Airtel","Vodafone","Jio","Idea","Airtel","Jio"],
     'Profit' : [1100,1120,1130,1150,1100,716]
}
df2 = pd.DataFrame(data1)
df2
```

```python
df3 = df.merge(df2, how='left', left_on=['Company'], right_on=['Organization'])
df3
```

```python
#Percentile calculation
data4 = {
    'Industry' : ['Retail','Finance','Sports','Food','Retail','Finance','Sports','Food','Retail','Finance','Sports','Food'],
    'Value': [1000,900,800,700,200,100,500,1500,15,10,19,23]
}
df7 = pd.DataFrame(data4)
df7
```

```python
percents =[90,75,50,25,10]
ind_unq = df7['Industry'].unique()

Result = []
data4_pd = []
for i in ind_unq:
  data4_pd += ([i]*5)
  unq = df7[(df7['Industry']==i)]
  for n in percents:
    cutoff = np.percentile(unq['Value'], n)
    Result.append(cutoff)

percent_pd = percents * len(ind_unq)

df8 ={'Industry1': data4_pd, 'Percents': percent_pd,'value1': Result}
wg_thresh = pd.DataFrame(df8)
wg_thresh.head(15)
```

```python
#Convert to wide format
wg_thresh1 = wg_thresh.pivot(index='Industry1', columns='Percents', values='value1')
wg_thresh1
```
