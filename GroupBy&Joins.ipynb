```python
# Python Basics - Collections in Python
# 1. List - Mutable, Indexed, Ordered, Allows Duplicates, Represented by []
# 2. Tuple - Immutable, Indexed, Ordered, Allows Duplicated, Represented by ()
# 3. Set - Mutable, Unorderd, Represented by {}
# 4. Dictionary - Key/Value, Mutable, Indexed, Ordered, Allows Duplicate, Represented by {}
```

```python
#Applying Groupby on dataframe
import pandas as pd
data= {
    'Company': ["Airtel","Vodafone","Jio","Idea","Airtel","Jio"],
     'Person': ["Agent1","Agent2","Agent3","Agent4","Agent5","Agent6"],
     'Sales' : [100,120,130,150,100,76]
}
df = pd.DataFrame(data)
df
```

```python
df1 = df.groupby(['Company','Person','Sales']).sum()
df1
```

```python
#Joins
data1= {
    'Organization': ["Airtel","Vodafone","Jio","Idea","Airtel","Jio"],
     'Profit' : [1100,1120,1130,1150,1100,716]
}
df2 = pd.DataFrame(data1)
df2
```

```python
df3 = df.merge(df2, how='left', left_on=['Company'], right_on=['Organization'])
df3
```
